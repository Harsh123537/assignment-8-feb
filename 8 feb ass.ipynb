Abstraction means to hide the implementation part of any application so that its complexity reduces.
For example there is a ticket booking system, there we only deals with booking the ticket and canceling the ticket but how does this process occur we don't care.

In abstraction the implementation part is being hide only functionality is being shown. On the other hand encapsulation means to compact the different methods and there data in one place.Abstraction is achieved  through encapsulation.
For example there is a ticket booking system, there we only deals with booking the ticket and canceling the ticket but how does this process occur we don't care this is abstraction on the other hand compacting the booking method, cancelling method and there data in one class is known as encapsulation.
 

abc module is known as abstract base class module.
It is used when any method of parent class compulsory need to be implement in child class.
For example 

```python
from abc import ABC,abstractmethod
class shape(ABC):
    @abstractmethod
    def printarea(self):
        pass
class rectangle(shape):
    def __init__(self,length,breath):
        self.length=length
        self.breath=breath
    def printarea(self):
        return self.length*self.breath
r=rectangle(7,8)
r.printarea()
```

Abstraction can be achieved by using abstract classes and interfaces.

An abstract class is a class, but not one you can create objects from directly. Its purpose is to define how other classes should look like, i.e. what methods and properties they are expected to have
